```python
 import pandas as pd
```

```python
new = pd.DataFrame({
    "date": ["5/1/2017","5/2/2017","5/3/2017"],
    "city":["new york","new york","new york"],
    "temperature":[32,35,23],
    "humidity":[56,45,24]
})
new
```

```python
new1 = pd.DataFrame({
    "date": ["5/1/2017","5/2/2017","5/3/2017"],
    "city":["itahari","itahari","itahari"],
    "temperature":[76,56,29],
    "humidity":[63,52,41]
})
new1
```

```python
new3 = pd.DataFrame({
    "date": ["5/1/2017","5/2/2017","5/3/2017"],
    "city":["berlin","berlin","berlin"],
    "temperature":[23,31,19],
    "humidity":[59,48,50]
})
new3
```

```python
df1 = pd.concat([new,new1,new3],ignore_index=True)
df1
```

```python
df1.to_csv("weather6.csv",index = False)
```
