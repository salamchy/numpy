```python
import pandas as pd
```

```python
nepal_weather = pd.DataFrame({
    "city": ['ith','brt','ktm'],
    "temperature":[45,76,34],
    "humidity": [80,56,37]
})
nepal_weather
```

```python
germany_weather = pd.DataFrame({
    "city": ['fkfnt','brln','hmbg'],
    "temperature":[21,18,31],
    "humidity": [60,46,37]   
})
germany_weather
```

**concat two dataframe**

```python
df = pd.concat([nepal_weather,germany_weather])
df
```

```python
df = pd.concat([nepal_weather,germany_weather],ignore_index = True)
df
```

```python
df = pd.concat([nepal_weather,germany_weather],keys = ["Nepal","Germany"])
df
```

```python
df.loc["Nepal"]
```

**Merge**

```python
df1 = pd.DataFrame({
    "city": ['ktm','ith','brt'],
    "temperature":[12,25,36]
})
df1
```

```python
df2 = pd.DataFrame({
    "city" : ['ith','ktm','brt'],
    "humidity":[65,75,59]
})
df2
```

```python
df3 = pd.merge(df1,df2,on = "city")
df3
```

```python
df4 = pd.DataFrame({
    "city": ['ktm','ith','brt','pkr'],
    "temperature":[12,25,36,19]
})
df4
```

```python
df5 = pd.DataFrame({
    "city" : ['ith','ktm','brt','dmk'],
    "humidity":[65,75,59,49]
})
df5
```

```python
df6 = pd.merge(df4,df5,on="city",how="outer")
df6
```

```python
df7 = pd.merge(df4,df5,on="city",how="left")
df7
```

```python
df8 = pd.merge(df4,df5,on="city",how="outer",indicator=True)
df8
```
