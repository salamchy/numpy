# matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [23,25,27,31,76,12,19]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, 'g+-')
```

```python
plt.plot(x,y, 'g+-.')
```

```python
plt.plot(x,y,color='green',alpha=0.5)
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,45]
```

```python
plt.plot(days,max_t)
```

```python
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
plt.xlabel("Days")
plt.ylabel("Temerature")
plt.title("weather")
plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label ="avg")
plt.legend()
plt.grid()
```

# Bar Chart

```python
import numpy as np
```

```python
company = ['GOGL','FB','AMZN','MCSFT']
revenue = [89,136,85,33]
profit = [45,80,32,12]
```

```python
ypos = np.arange(len(company))
ypos
```

```python
plt.xticks(ypos,company)
plt.xlabel('days')
plt.ylabel('revenue')
plt.title('Revenue of Company')
plt.bar(ypos,revenue,label = "revenue")
plt.bar(ypos,profit,label="profit")
plt.legend()
```

```python
company = ['GOGL','FB','AMZN','MCSFT']
revenue = [89,136,85,33]
profit = [45,80,32,12]
```

```python
ypos = np.arange(len(company))
ypos
```

```python
plt.xlabel('days')
plt.ylabel('revenue')
plt.title('Revenue of Company')
plt.bar(ypos-0.2,revenue,width = 0.4,label = "revenue")
plt.bar(ypos+0.2,profit,width = 0.4,label="profit")
plt.legend()
```

# histogram

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar, bins =3,rwidth=0.95,color='green')
```

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.xlabel('sugar range')
plt.ylabel('no of patients')
plt.title('Blood Sugar Analysis ')
plt.hist([blood_sugar_men,blood_sugar_women], bins =3, rwidth=0.95, color=['green','orange'],label =['men','women'])
plt.legend()
```

# Pie-Chart

```python
exp_vals = [1555,888,300,415,350]
exp_labels = ['phone_bill','book','pen','car','others']

plt.axis('equal')
plt.pie(exp_vals, labels = exp_labels, radius= 1.2, autopct = '%0.1f%%',startangle= 180)
plt.show()
```

```python
exp_vals = [155,600,360,215,350]
exp_labels = ['phone_bill','book','pen','car','others']

plt.axis('equal')
plt.pie(exp_vals, labels = exp_labels, radius= 1.2, autopct = '%0.1f%%',explode=[0,0,0.1,0.1,0])
plt.show()
plt.savefig("chart.png",bbox_inches="tight")
```
